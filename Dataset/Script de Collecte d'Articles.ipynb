```python
# Clé API GNews (remplacez par votre propre clé)
gnews_api_key = "58c7b665f697009817d8a60e79fe0da0"

# Domaines sélectionnés
categories = ["sports", "technology", "politics", "entertainment", "business"]

# Langue
language = "en"

# Fonction pour créer un dossier si nécessaire
def create_folder(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

# Fonction pour sauvegarder un article dans un fichier séparé
def save_article(category, description, index):
    folder_name = f"articles_final_gnews/{category}"
    create_folder(folder_name)
    file_path = os.path.join(folder_name, f"article_{index + 1}.txt")
    # Vérifier si la description est None et remplacer par une chaîne vide
    description = description if description else "No description available"
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(description)

# Fonction pour collecter les articles depuis GNews API
def fetch_gnews_api():
    url = "https://gnews.io/api/v4/search"
    for category in categories:
        page = 1
        total_articles = 0
        while total_articles < 100:  # Limite à 100 articles par catégorie (ou selon votre quota)
            params = {
                "q": category,
                "lang": language,
                "token": gnews_api_key,
                "page": page,
                "max": 20  # Nombre d'articles par requête (maximum 20 articles par requête avec le plan gratuit)
            }
            response = requests.get(url, params=params)
            
            if response.status_code == 200:
                data = response.json()
                articles = data.get("articles", [])
                for index, article in enumerate(articles):
                    description = article.get("description") or "No description available"
                    save_article(category, description, total_articles + index)
                total_articles += len(articles)
                page += 1  # Passer à la page suivante
            else:
                print(f"Erreur avec GNews API pour {category}: {response.status_code}")
                break  # Quitter la boucle en cas d'erreur

            # Pause pour éviter de dépasser le quota
            time.sleep(5)  # Pause de 5 secondes entre chaque requête

# Appeler la fonction pour collecter les articles
fetch_gnews_api()

print("Articles collectés et organisés par catégorie dans des fichiers séparés.")
```

```python
# Clé API Mediastack (remplacez par votre propre clé)
mediastack_api_key = "f58013475dd09f0104b2836fc971a0e3"

# Domaines sélectionnés
categories = ["sports", "technology", "politics", "entertainment", "business"]

# Langue
language = "en"

# Fonction pour créer un dossier si nécessaire
def create_folder(folder_name):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

# Fonction pour sauvegarder un article dans un fichier séparé
def save_article(category, description, index):
    folder_name = f"articles_final_mediastack/{category}"
    create_folder(folder_name)
    file_path = os.path.join(folder_name, f"article_{index + 1}.txt")
    # Vérifier si la description est None et remplacer par une chaîne vide
    description = description if description else "No description available"
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(description)

# Fonction pour collecter les articles depuis Mediastack API
def fetch_mediastack_api():
    url = "http://api.mediastack.com/v1/news"
    for category in categories:
        page = 1
        total_articles = 0
        while total_articles < 100:  # Limite à 100 articles par catégorie (ou selon votre quota)
            params = {
                "access_key": mediastack_api_key,
                "categories": category,
                "languages": language,
                "page": page,
                "limit": 20  # Nombre d'articles par requête (maximum 20 articles par requête avec le plan gratuit)
            }
            response = requests.get(url, params=params)
            
            if response.status_code == 200:
                data = response.json()
                articles = data.get("data", [])
                for index, article in enumerate(articles):
                    description = article.get("description") 
                    save_article(category, description, total_articles + index)
                total_articles += len(articles)
                page += 1  # Passer à la page suivante
            else:
                print(f"Erreur avec Mediastack API pour {category}: {response.status_code}")
                break  # Quitter la boucle en cas d'erreur

            # Pause pour éviter de dépasser le quota
            time.sleep(5)  # Pause de 5 secondes entre chaque requête

# Appeler la fonction pour collecter les articles
fetch_mediastack_api()

print("Articles collectés et organisés par catégorie dans des fichiers séparés.")
```
